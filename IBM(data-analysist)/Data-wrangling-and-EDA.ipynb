### Analysis of Variance ANOVA 

`• Why do we perform ANOVA?` <br>
• Finding correlation between different groups of a categorical variable.<br>
`• What we obtain from ANOVA?`<br>
• F-test score : variation between sample group means divided by variation
within sample group <br>
• p-value : confidence degree<br>

```python
fig = plt.figure(figsize =(30, 40))
plt.bar(df['make'],df['price'],width=0.8)
plt.xlabel("Makes")
plt.ylabel("Prices")
plt.title("Price vs make barchat")
```

```python
df_anova = df[['make','price']]
grouped_anova = df_anova.groupby(['make'])
```

```python
from scipy import stats
import seaborn as sns
```

```python
anova_result = stats.f_oneway(grouped_anova.get_group('honda')['price'],grouped_anova.get_group('isuzu')['price'])
anova_result
```

```python
anova_result = stats.f_oneway(grouped_anova.get_group('honda')['price'],grouped_anova.get_group('jaguar')['price'])
anova_result
```

```python
sns.regplot(x='engine-size',y='price',data=df)
plt.ylim(0,)
df['engine-size'].corr(df['price'])
```
